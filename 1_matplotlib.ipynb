```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

### Matplotlib Basics

```python
# Step 1: Create some data
x = [1, 2, 3, 4, 5]  # X-coordinates (horizontal)
y = [2, 5, 3, 8, 7]  # Y-coordinates (vertical)

# Step 2: Create the plot
plt.figure(figsize=(6, 3))
plt.plot(x, y)
# Step 3: Add labels and title
plt.xlabel('Days')           # Label for X-axis
plt.ylabel('Temperature')    # Label for Y-axis
plt.title('Temperature Over 5 Days')  # Title at the top
plt.plot(x, y, marker='o')
# Step 4: Show the plot
plt.show()
```

### Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```
